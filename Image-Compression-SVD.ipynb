```python
import numpy as np
import numpy.linalg as la
import imageio as io
import matplotlib.pyplot as plt
import scipy.misc as msc
```

```python
F = msc.face(gray=True)
```

```python
plt.imshow(F, cmap='gray')
```

```python
F1 = F[200:400 , 500:700]
```

```python
plt.imshow(F1, cmap='gray')
```

```python
U,S,VT = la.svd(F)
```

```python
r = len(S)
```

```python
def Ftest(x):
  return U[:,:x].dot(np.diag(S[:x])).dot(VT[:x])
```

```python
def store(s):
  G = Ftest(s)
  U,S,VT = la.svd(G)
  return (len(U)*s)+s+(len(VT)*s)
```

```python
p0 = store(r)
```

```python
def acc(p):
  return p/p0
```

```python
plt.imshow(Ftest(r), cmap='gray')
print(store(r))
print(acc(p0))
```

```python
plt.imshow(Ftest(500), cmap='gray')
p500 = store(500)
print(p500)
print(acc(p500))
```
